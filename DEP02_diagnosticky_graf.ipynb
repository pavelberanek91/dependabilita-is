```python
from math import factorial

def rekurentni_soucin_souctu(n, ik, iterators, ck, matice):
    if ik > 0:
        for iki in range(n):
            iterators[ik-1] = iki
            rekurentni_soucin_souctu(n, ik-1, iterators, ck, matice)
    else:
        if not(len(set(iterators)) != len(iterators) and len(iterators) != 1):
            ck[0] += 1 if all([any([matice[iki][j] == 1 for iki in iterators]) for j in range(len(matice))]) else 0

def charakteristicka_cisla(matice):
    n = len(matice)
    cks = []
    for ik in range(1, n+1):
        ck = [0]
        rekurentni_soucin_souctu(n, ik, iterators = [0]*ik, ck=ck, matice=matice)
        cks.append(ck[0]/factorial(ik))
    return cks
    
def pravdepodobnost_spravne_diagnostiky(matice, Pm):
    n = len(matice)
    cks = charakteristicka_cisla(matice)
    PSD = 0
    for k in range(1, n+1):
        PSD += cks[k-1]*Pm**(n-k)*(1-Pm)**k
    return PSD

    
```
